```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
df = pd.read_csv('calibration.csv')
df=df[df["0"]!=0]
print(df.to_string()) 
```

```python
xcorr=(np.max(df["0"])+np.min(df["0"]))/2
ycorr=(np.max(df["0.1"])+np.min(df["0.1"]))/2
zcorr=(np.max(df["0.2"])+np.min(df["0.2"]))/2
print(xcorr)
```

```python
plt.plot(df["0"]-xcorr,df["0.1"]-ycorr,".")
plt.plot(0,0,"o")
plt.show()
plt.plot(df["0"]-xcorr,df["0.2"]-zcorr,".")
plt.plot(0,0,"o")
plt.show()
plt.plot(df["0.2"]-zcorr,df["0.1"]-ycorr,".")
plt.plot(0,0,"o")

```

```python
json_str ='{'+ f'"xcorr":{xcorr}, "ycorr": {ycorr}, "zcorr": {zcorr}'+',"sf":1.5}'
print(json_str)
data = json.loads(json_str)
print(data)
```
